```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
```

Import dependencies

```
housing = fetch_california_housing()
```

Import the dataset

```
df = pd.DataFrame(housing.data, columns=housing.feature_names)
sns.relplot(x="AveOccup", y ="Population", data=df)
```

```
sns.relplot(x="AveOccup", y ="Population", hue="HouseAge", data=df)
```

```
sns.relplot(x="AveOccup", y ="Population", data=df, kind = "line")
```

```
sns.catplot(x="AveOccup", y ="Population", data=df, kind = "violin")
```
